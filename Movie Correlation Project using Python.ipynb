# Importing Libraries

```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Adjusts the configuration of the plots that are created.
```

# Reading the data

```python
df = pd.read_csv('movies.csv')
```

# Viewing the data 

```python
df
```

```python
df.shape
```

```python
df.head()
```

# Checking for any missing data

```python
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

# Checking data types of columns

```python
df.dtypes
```

# Filling Missing Data

```python
df = df.fillna(0)
```

# Changing datatype

```python
df['budget'] = df['budget'].astype('int64') 
df['gross'] = df['gross'].astype('int64') 
```

# Checking for outliners

```python
df.boxplot(column=['gross'])
```

# Dropping duplicates 

```python
df.drop_duplicates()
```

# Sorting the data

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

# Scatter Plot with Budget vs Revenue

```python
plt.scatter(x=df['budget'],y=df['gross'])
plt.xlabel('Budget in USD million (in 100s)')
plt.ylabel('Gross Earnings in USD billion')
plt.title('Scatter Plot - Budget Vs Gross Earnings')
plt.show()
```

```python
df.head()
```

# Regression Plot - Budget Vs Gross Earnings

```python
sns.regplot(x = 'budget',y = 'gross',data=df,scatter_kws={'color':'red'},line_kws={'color':'blue'})
plt.xlabel('Budget in USD million (in 100s)')
plt.ylabel('Gross Earnings in USD billion')
plt.title('Regression Plot - Budget Vs Gross Earnings')
```
